```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
```

```python
df[0:12]
```

```python
X=df.drop('target',axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=20)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
import pandas as pd
df=pd.read_csv("Csv files/housing_data.csv",parse_dates=['date'])
df.head()
```

```python
df['Month']=df['date'].dt.month
```

```python
df['Year']=df['date'].dt.year
```

```python
df
```

```python
df.isna().sum()
```

```python
X=df.drop(columns=['id','date','price'])
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=64)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor()
reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred_train=reg.predict(X_train)
mean_absolute_error(y_train,y_pred_train)
```

```python
y_pred_test=reg.predict(X_test)
mean_absolute_error(y_test,y_pred_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[120,125,130],
            'max_depth':[16,18,22],
            'max_features':[9,10,11],
            'max_leaf_nodes':[84,86,88]
           }
reg1=RandomForestRegressor()
reg2=GridSearchCV(reg1,parameters)
            
```

```python
reg2.fit(X_train,y_train)
```

```python
y_pred_train=reg1.predict(X_train)
mean_absolute_error(y_train,y_pred_train)
```

```python
y_pred_test=reg1.predict(X_test)
mean_absolute_error(y_test,y_pred_test)
```
